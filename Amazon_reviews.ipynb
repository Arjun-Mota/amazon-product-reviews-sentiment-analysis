# Identifying Most Reviewed Product from Amazon DataSet of 35000 approx. reviews
We've dataset from amazon with reviews of many products, <br>out of them we need to know which are the most reviewed products?<br>
Let's look at dataset and see how we can do this!

```python
import pandas as pd
df=pd.read_csv("1429_1.csv")
df.head()
```

### Data Lookup
After looking at data from dataset, it's look like there are some NaN values in it which we don't require

```python
df.shape
```

```python
df.tail()
```

## Changing Column Names
As column names are lengthy and not informative, so we'll give some suitable names to them

```python
df.columns = ['id', 'name', 'asins', 'brand', 'categories', 'keys', 'manufacturer',
       'date', 'dateAdded', 'dateSeen',
       'isPurchase', 'isRecommended', 'reviewsId',
       'numHelpful', 'rating', 'sourceURLs',
       'reviewText', 'reviewTitle', 'city',
       'userProvince', 'username']
df.head()
```

## Handling Null values

```python
df.reviewsId.dtype
```

```python
df.dtypes
```

```python
df.reviewsId.fillna(0.0)
df.head()
```

## Removal of unnecessary columns
To identify most reviewed products, we don't required few columns, therefore we'll remove those columns from dataset

```python
df.columns
```

```python
df=df.drop('keys',1)
df.drop('sourceURLs', 1, inplace=True)
df.drop(['dateAdded','dateSeen'], 1, inplace=True)
df.head()
```

```python
df.isPurchase.fillna(False,inplace=True)
df.reviewsId.fillna("",inplace=True)
df.city.fillna("",inplace=True)
df.userProvince.fillna("",inplace=True)
df.head()
```

## DataSet Statistics Lookup

```python
df.describe(include='object')
```

Removal of rows which are affecting our estimation

```python
df.dropna(subset=['name'],inplace=True)
df.describe(include='object')
```
